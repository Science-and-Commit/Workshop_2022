## setup.py

<font size=3> setup.py es un script de construcción para setuptools. Le dice a setuptools sobre tu paquete (su nombre y versión) al igual que los archivos de código por incluir.

Abre setup.py e ingresa el siguiente contenido. Puedes actualizar tu paquete para que tenga tu nombre de usuario, esto asegura que tienes un paquete de nombre único para así no tener conflicto con otros paquetes que hayan sido subidos. <font>

```python
import setuptools

with open("README.md", "r") as fh:
    long_description = fh.read()

setuptools.setup(
    name="Mi-paquete-Mi-Usuario", # Replace with your own username
    version="0.0.1",
    author="Autor de Ejemplo",
    author_email="autor@ejemplo.com",
    description="Un paquete de ejemplo",
    long_description=long_description,
    long_description_content_type="text/markdown",
    url="https://github.com/pypa/sampleproject",
    packages=setuptools.find_packages(),
    classifiers=[
        "Programming Language :: Python :: 3",
        "License :: OSI Approved :: MIT License",
        "Operating System :: OS Independent",
    ],
    python_requires='>=3.6',
)
```

<font size=3> setup() toma varios argumentos. Este ejemplo ocupa un set mínimo de ellos: <font>



- $\textbf{name}$ Es el nombre de distribución de tu paquete. Puede ser cualquier nombre siempre y cuando contenga letras, números, _ y -. Además, no debe estar ocupado en pypi.org. Se recomienda añadir el usuario al nombre del paquete para asegurarse que sea un nombre único.

- $\textbf{version}$ Es la versión del paquete. Para más información de versiones ir a https://www.python.org/dev/peps/pep-0440/#abstract

- $\textbf{author}$ y $\textbf{author_email}$ son usados para identificar el autor del paquete.

- $\textbf{description}$ Es un resumen breve del paquete.

- $\textbf{long_description}$ Es una descripción detallada del paquete. Esto se muestra en el detalle del paquete en el Python Package Index. En este caso, long_description se carga desde README.md, que es un patrón común.

- $\textbf{long_description_content_type}$ Le dice al índice, qué tipo de markup es usado para long description. En este caso, es Markdown.

- $\textbf{url}$ Es el url para la página principal del proyecto. Para varios proyectos, éste será el link a GitHub, GitLab, Bitbucket, u otras páginas similares. 

- $\textbf{packages}$ Es la lista de todos los paquetes por importar en Python que deben ser incluidas en el paquete de distribución. En vez de enlistarlo manualmente, se puede utilizar find_packages() que automáticamente descubrirá todos los paquetes y sub-paquetes.

- $\textbf{classifiers}$ Le da al índice y a pip metadata adicional para tu paquete. En este caso, el paquete es sólo compatible con Python 3, está bajo la licencia de MIT, y es independiente de OS. Siempre se debería incluir en qué versión(es) de Python funciona el paquete, bajo qué licencia está disponible tu paquete, en qué sistemas operativos funciona el paquete. Para una lista completa de classifiers, visita https://pypi.org/classifiers/.

Hay muchas más variables que las mencionadas aquí. Visita 'Packaging and distribution of Packages' https://packaging.python.org/guides/distributing-packages-using-setuptools/ para más detalles.



## Archivos de Distribución 

<font size=3> Ahora es momento de crear el Archivo de Distribución, que es el archivo que un usuario descargará de internet e instalará. Estos son archivos que son subidos al índice del Paquete y pueden ser instalados con $\textbf{pip}$.  
    $\\$
    Asegúrate de tener instalados las últimas versiones de $\textbf{setuptools}$ y $\textbf{wheel}$ <font>

```python
python3 -m pip install --user --upgrade setuptools wheel ## macOS/Linux
py -m pip install --upgrade build ## Windows
```

<font size=3> Ahora compila este comando en el mismo directorio donde se encuentra $\textbf{setup.py}$ <font>

```python
python3 setup.py sdist bdist_wheel ## macOS/Linux
py -m build ## Windows
```

<font size=3> Este comando da como output mucho texto, y una vez completado se deberían generar dos archivos en el directorio<font>

```python
Mi-paquete-Mi-Usuario-0.0.1-py3-none-any.whl
Mi-paquete-Mi-Usuario-0.0.1.tar.gz
```

### ¡Subiendo el archivo de distribución!

Finalmente, es hora de subir tu paquete al Índice de paquetes de Python :) 

Lo primero que hay que hacer es registrarse en una cuenta en Test PyPI. Test PyPI es una instancia aparte del del paquete de índices, destinado al testeo y la experimentación. Para registrarse a una cuenta, anda a https://test.pypi.org/account/register/ y completa los pasos de esa página. También necesitarás verificar el mail antes de ser capaz de subir cualquier paquete. Para más detalles en Test PyPI, anda a https://packaging.python.org/guides/using-testpypi/.

Ahora crearás un token de PyPI API, para subir seguramente tu proyecto..

Anda a https://test.pypi.org/manage/account/#api-tokens y crea un nuevo token de API; no limites su alcance a un proyecto particular, ya que estás creando un nuevo proyecto 

No cierres esta página hasta que hayas copiado y guardado el nuevo API token; no verás ese token otra vez. 

Ahora que estas registrade, puedes usar twine (https://packaging.python.org/key_projects/#twine) para subir los paquetes de distribución. Tendrás que instalar Twine:

```python
python3 -m pip install --user --upgrade twine ## macOS/Linux
py -m pip install --user --upgrade twine ## Windows
```

Una vez instalado, corre Twine para subir todos los archivos bajo dist:

```python
python3 -m twine upload --repository testpypi dist/* ## macOS/Linux
py -m twine upload --repository testpypi dist/* ## Windows
```

Te pedirá un usuario y contraseña. Puedes ocupar la cuenta y contraseña que te acabas de crear o usar un __token__ para el usuario y el valor del token con el prefijo pypi- para la contraseña. 

Luego que el comando este completo, deberías ver un output similar a esto:

```python
Uploading distributions to https://test.pypi.org/legacy/
Enter your username: [tu usuario]
Enter your password:
Uploading example_pkg_YOUR_USERNAME_HERE-0.0.1-py3-none-any.whl
100%|█████████████████████| 4.65k/4.65k [00:01<00:00, 2.88kB/s]
Uploading example_pkg_YOUR_USERNAME_HERE-0.0.1.tar.gz
100%|█████████████████████| 4.25k/4.25k [00:01<00:00, 3.05kB/s]
```

Una vez subido, tu paquete deberia ser visible en TestPyPI, por ejemplo, https://test.pypi.org/project/Mi-paquete-mi-usuario

Para subir tu paquete a PyPI, ocupa el comando:

```python
twine upload dist/* ## macOS/Linux
```

Donde también te pedirá un usuario y contraseña.
