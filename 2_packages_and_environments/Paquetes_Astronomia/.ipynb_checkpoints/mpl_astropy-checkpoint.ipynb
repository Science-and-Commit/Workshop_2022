# A. Matplotlib.pyplot

### Tips generales:
**No sobrediseñe su código para graficar.** Esto significa que no necesita agregar un millón de palabras clave anticipando cada ajuste posible que sus usuarios deseen realizar. Concéntrese en facilitar que sus usuarios modifiquen el código ellos mismos. Asegúrese de devolver los objetos `Figura` y `Axes` (más sobre esto a continuación) a su usuario para que pueda modificar fácilmente los gráficos.

**Aplica los principios de un buen diseño.** Esto significa usar paletas de colores con mucho contraste, hacer que tus fuentes sean grandes y fáciles de leer, y asegurarte de que tus gráficos sean aptos para daltónicos. (Existen varios simuladores de daltonismo geniales, por ejemplo [este](https://www.color-blindness.com/coblis-color-blindness-simulator/).) También es bueno diferenciar los elementos de la trama en más de de una manera (por ejemplo: una línea es negra y discontinua, otra es roja y continua).

### Este tutorial parte aca:

`matplotlib.pyplot` se basa en el concepto de *estado cambiante*. Puede pensar en el *estado* de cualquier entorno (su shell bash, por ejemplo, o el indicador de iPython) como una lista de variables y comandos precargados que controlan los comportamientos predeterminados. Algunas de estas variables son **mutables**, lo que significa que puede cambiarlas según sus necesidades de programación. Cambiar las variables de estado mutables solo significa actualizar el comportamiento predeterminado de su entorno.

Un excelente ejemplo de este concepto es el objeto `matplotlib.Rcparams`, que es como un diccionario de atributos que controlan los comportamientos predeterminados de los diagramas de matplotlib. Usemos el método `update()` de este objeto para cambiar el tamaño de fuente predeterminado:

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt

# hacemos el plot
plt.figure()
plt.title('El tamaño de fuente por default es {}.'.format(rcParams['font.size']))
plt.xlabel('foo'); plt.ylabel('bar')

# actualizar el estado actual
rcParams.update({'font.size': 22})

# hacemos otro plot
plt.figure()
plt.title('Ahora, el nuevo tamaño default es {}.'.format(rcParams['font.size']))
plt.xlabel('foo'); plt.ylabel('bar', fontsize=10)

# actualizar el estado actual nuevamente
rcParams.update({'font.size': 12})
```

Como puede ver, `matplotlib` adopta el tamaño de fuente predeterminado al verificar el valor *actual* del parámetro de estado `rcParams['font.size']`, pero este se sobrescribe si se le da un nuevo valor en al graficar (en este caso, el tamaño de fuente de la etiqueta del eje y).

La "unidad" básica de `matplotlib.pyplot` es [el objeto `Figura`](https://matplotlib.org/3.2.1/api/_as_genmatplotlib.figure.Figure.html#matplotlib.figure.Figure), que actúa como un contenedor para todos los demás elementos. Veamos algunas formas diferentes de crear un objeto `Figura`:

```python
import numpy as np

# Datos de ejemplo
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])
y_error = np.array([5, 1.2, 2, 5])

fig = plt.figure() # inicializamos el objeto `Figure`
# graficamos en el objeto `Figure`
plt.errorbar(x, y, y_error, ls='', fmt="*")  

print(type(fig))
```

La figura actual cambia (es decir, el *estado* se actualiza) cuando inicializa un nuevo objeto `Figura`. A menos que especifique un objeto `Figura` en particular, las funciones `plt.___()` siempre se agregarán al objeto `Figura` actual. Esa es una buena característica: facilita agregar múltiples tipos de gráficos en la misma figura.

Veamos otra forma de inicializar un nuevo objeto `Figura`:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

ax[0].plot(x, y)
ax[1].plot(x, y)
ax[1].set_yscale('log')

print('fig type: {}'.format(type(fig)))
print('ax type: {}'.format(type(ax)))
print('ax[0] type: {}'.format(type(ax[0])))
```

Aquí hemos encontrado otro tipo de objeto: [el objeto `Axes`](https://matplotlib.org/api/axes_api.html#matplotlib.axes.Axes). Estos objetos son contenedores para todos los elementos del gráfico *en un panel de figura en particular*. Los `ejes` son atributos de los objetos `Figura`:

```python
fig.axes[0].set_xlim(1, 3)
fig
```

Dos métodos útiles a tener en cuenta son `plt.gca()` y `plt.sca()` (obtener/establecer `Ejes` actuales). Veamos cómo funcionan:

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

mystery_axes = plt.gca()              # accedemos a los Axes actuales
mystery_axes.scatter(x, y, color='k') # actua en `mystery_axes`

plt.sca(ax[0,0])        # set current Axes
plt.plot(x, y, ls='-.') # acts on current Axes
plt.tight_layout()      # prettify
```

Una vez más, este tutorial está lejos de ser exhaustivo. Consulte una lista más completa de tutoriales de `astropy` [aquí] (http://learn.astropy.org/tutorials.html).

A continuación se muestran algunos ejemplos de algunas características particularmente útiles de `astropy`.

##### Ejemplo 1
Una de las mejores propiedades de `astropy` es su habilidad para manejar unidades y constantes fundamentales. 

Calculemos la velocidad orbital de la Tierra:

```python
import numpy as np
from astropy import constants as cst, units as u 

v_Earth = np.sqrt(cst.G * cst.M_sun / (1 * u.au))

print(v_Earth)
```

Bueno, esas unidades no son muy útiles. Vamos a convertirlas a unidades físicas:

```python
v_Earth_cgs = v_Earth.cgs
v_Earth_km_s = v_Earth.to(u.km/u.s)

print("Velocidad orbital de la tierra: {:.1f}.".format(v_Earth_cgs))
print("En otras unidades: {:.1f}.".format(v_Earth_km_s))
```

##### Ejemplo 2
A continuación, hagamos un ejemplo con la ruina de la existencia de todo astrónomo: el tiempo.

```python
from astropy.time import Time 

my_birthday = Time('1994-11-04', format='iso')

print('Un año con decimales: {}.'.format(my_birthday.decimalyear))
print('En JD: {}.'.format(my_birthday.jd))
print('En MJD: {}.'.format(my_birthday.mjd))
```

##### Ejemplo 3
A continuación, un ejemplo rápido con la segunda peor pesadilla de la existencia de todo astrónomo: los sistemas de coordenadas. `astropy` tiene un excelente módulo de `coordenadas` que permite una fácil traducción entre sistemas de coordenadas.

```python
from astropy.coordinates import SkyCoord
from astropy import units as u

# input coordinates as hour angle and degree
my_coordinate = SkyCoord('00:42.5 +41:12', unit=(u.hourangle, u.deg))

# access right ascension (RA)
print(my_coordinate)

# convert to galactocentric coordinates
print(my_coordinate.galactic)
```
